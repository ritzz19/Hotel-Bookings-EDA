# Bivariant Analysis

```python
fig, axes = plt.subplots(7, 2, figsize=(16, 24))
axes = [ax for axes_row in axes for ax in axes_row]

for i, c in enumerate(data[cat_cols]):
    fltr = data["hotel"] == "City Hotel"
    vc_a = data[fltr][c].value_counts(normalize=True).reset_index().rename({'index' : c, c: 'count'}, axis=1)

    vc_b = data[~fltr][c].value_counts(normalize=True).reset_index().rename({'index' : c, c: 'count'}, axis=1)

    vc_a["hotel"] = "City Hotel"
    vc_b["hotel"] = "Resort Hotel"

    df = pd.concat([vc_a, vc_b]).reset_index(drop = True)

    _ = sns.barplot(y = c, x = 'count', data =df , hue="hotel", ax=axes[i])
    
    
_ = plt.tight_layout()
```

#### 1.Country we will plot separately as we will take only top 10
#### 2. 'is_canceled','is_repeated_guest','same_room' are not proper with normalize=True therefore we will plot it later

```python
fig, axes = plt.subplots(1, 1, figsize=(16, 10))

fltr = data["hotel"] == "City Hotel"
vc_a = data[fltr]['country'].value_counts(normalize=True).head(10).reset_index().rename({'index' : 'country', 'country': 'count'}, axis=1)

vc_b = data[~fltr]['country'].value_counts(normalize=True).head(10).reset_index().rename({'index' : 'country', 'country': 'count'}, axis=1)

vc_a["hotel"] = "City Hotel"
vc_b["hotel"] = "Resort Hotel"

df = pd.concat([vc_a, vc_b]).reset_index(drop = True)

_ = sns.barplot(y = 'country', x = 'count', data =df , hue="hotel")
    
    
_ = plt.tight_layout()
```

## Bivariant conclusion:
### 1. From reservation_status it is clear that though City Hotel is 64% but Resort Hotel has more number of guests and City Hotel has more number of Cancellation as compare to Resort Hotel
### 2. From market_segment and distribution_channel we can observe that City Hotel is more into 'Online TA' and Resort Hotel is more into 'Direct' segment.
### 3. From meal we can observe that City Hotel provide 'BB' and Resort Hotel provide 'HB'.
### 4. City Hotel is more non-refund as compare to Resort Hotel
### 5. Most of the guests are from PRT and they happen to choose Resort Hotel more than City Hotel
